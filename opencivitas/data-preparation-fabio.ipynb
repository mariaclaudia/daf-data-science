## Merge dei Dati

```python
df_spese_comuni_merge = pd.merge(
    df_spesa_comuni, 
    df_comuni, 
    how='left',
    right_on='Codice Nazionale', 
    left_on='COMUNE_CAT_COD')
df_spese_comuni_merge.shape
```

## Merge Dati con [DatiIRPEF](http://localhost:8888/notebooks/mise_aiuti/DAF_MEF_DatiIRPEF.ipynb)

```python
df_irpef_comuni = pd.read_csv('../mise_aiuti/data/redditi_irpef.csv')
df_irpef_comuni = df_irpef_comuni.drop(df_irpef_comuni.columns[0], axis=1)
```

## Add Total Population that pay taxes

filter the columns referring to the frequency of the people that pay taxes in the city. The sum of these values represent the total number of city citizens.

```python
frequency_cols = [
 'Reddito complessivo da 0 a 10000 euro - Frequenza',
 'Reddito complessivo da 10000 a 15000 euro - Frequenza',
 'Reddito complessivo da 15000 a 26000 euro - Frequenza',
 'Reddito complessivo da 26000 a 55000 euro - Frequenza',
 'Reddito complessivo da 55000 a 75000 euro - Frequenza',
 'Reddito complessivo da 75000 a 120000 euro - Frequenza',
 'Reddito complessivo minore o uguale a zero euro - Frequenza',
 'Reddito complessivo oltre 120000 euro - Frequenza',
]
```

```python
tot_citizens = df_final[frequency_cols].sum(axis=1)
```

```python
df_final= df_final.drop('tot_abitanti', axis=1)
```

```python
df_final.insert(df_final.shape[1], 'tot_abitanti', tot_citizens)
```

```python
df_br = df_final[(df_final['Sigla Provincia_y'] == 'RM')]
df_br = df_br.dropna(axis=1, how='all')
```
