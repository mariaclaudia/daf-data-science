## Columns Description

-  ANNO: anno rif
-  COMUNE_CAT_COD: cod comune
-  CAT_DET_COD_1: coefficiente popolazione residente ottenuta applicando modello di regressione definito nella analisi
-  CAT_DET_COD_2: coefficiente economia locale ottenuto applicando modello di regressione definito nella analisi
-  CAT_DET_COD_3: coefficiente prezzo degli input (vedi img sotto) ottenuto applicando modello di regressione definito nella analisi
-  CAT_DET_COD_4: coefficiente valore degli immobili ottenuto applicando modello di regressione definito nella analisi
-  CAT_DET_COD_5: coefficiente demografia ottenuto applicando modello di regressione definito nella analisi
-  COEFF_TEORICO: coefficiente del fabbisogno standard teorico per il comune
-  IND1: spesa storica del comune del servizio riscossione tributi per abitante in euro
-  IND3: % spesa storica per riscossione rispetto a fabbisogno
-  IND4: % Livello servizi offerti vs livello servizi standard(%)
-  IND5: numero di bollette o avvisi preventivi per famiglia(numero)
-  IND6: numero di avvisi di accertamento recupero evasione per famiglia(numero)
-  IND7: numero di cartelle di pagamento emesse per famiglia
-  IND8: numero di certificazioni fiscali per 1000 abitanti
-  IND9: numero di autorizzazioni per 1000 abitanti (si correla alla variabile sopra ???)
-  IND10: numero ore di apertura uffici al pubblico
-  IND11: percentuale velocita' riscossione tasse
-  LQP_COD_1: differenza tra fabbisogno standard e capacità fiscale,
-  LQP_COD_2: 
-  LQP_COD_3: comune non valutabile (0/1)
-  LQP_COD_4: Flag percentile anomalo (differenza % minore del 5^ percentile o maggiore del 95^ percentile)
-  LQP_COD_5: Posizione del Comune rispetto alla spesa
-  LQP_COD_6: Posizione del Comune rispetto al livello dei servizi
-  LQP_COD_S: Misura (da 0 a 10) della capacità del Comune di soddisfare la domanda di servizi espressa dai cittadini, tenendo conto della spesa e dei servizi offerti rispetto allo standard

### COEFF_TEORICO

Fabbisogno standard teorico, rappresenta il fabbisogno per ogni comune calcolato in base ai dati osservati fino alla data del 2010

```python
print('coeff teorico rispetto a fabbisogno storico al 2010 ', sum(df_serv_entrate.loc[df_serv_entrate['ANNO'] == 2010].COEFF_TEORICO))
print('coeff teorico rispetto a fabbisogno storico al 2013 ', sum(df_serv_entrate.loc[df_serv_entrate['ANNO'] == 2013].COEFF_TEORICO))
```
